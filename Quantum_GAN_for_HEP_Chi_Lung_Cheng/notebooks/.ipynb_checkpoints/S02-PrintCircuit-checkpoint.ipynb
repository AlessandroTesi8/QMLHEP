```python
def create_generator(data_circuit, model_circuit, readout, latent_dim):
    generator = Sequential()
    # input layer for feeding classical data
    input_layer = Input(shape=(latent_dim), dtype=tf.float32)
    pqc_layer = PQC(model_circuit, data_circuit, readout)
    generator.add(input_layer)
    generator.add(pqc_layer)
    return generator
```
