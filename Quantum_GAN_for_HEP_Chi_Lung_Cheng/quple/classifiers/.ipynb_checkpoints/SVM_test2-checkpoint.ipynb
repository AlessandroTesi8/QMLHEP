```python
import numpy as np
from pdb import set_trace
#qiskit modules
from qiskit.aqua import QuantumInstance, aqua_globals
from qiskit import Aer
import quple
from quple.classifiers.qsvm import QSVM
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from quple.data_encoding import  GeneralPauliEncoding, EncodingCircuit
from quple import ParameterisedCircuit, PauliBlock
from quple.qiskit_interface.tools import get_qiskit_state_vectors
from quple.qiskit_interface.feature_maps import FeatureMapDev
from quple.data_encoding.encoding_maps import trial_0_qiskit, trial_2
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score,roc_auc_score
from quple.components.data_preparation import prepare_train_val_test
import tensorflow as tf
random_seed = 10598
```

```python
# data preparation
data = np.load('/afs/cern.ch/work/c/chlcheng/Repository/quple/data/ttH_padded.npy')
x = data[:, :-1]
y = data[:, -1]
n_trials = 10
depth = 2
n_qubit = 10
n_event = 800
```

```python
aucs = np.zeros(n_trials)
for i in range(n_trials):
    encoding_circuit =EncodingCircuit(feature_dimension=n_qubit, 
                                       copies=depth,
                                       rotation_blocks=['H',PauliBlock('Z', trial_2(1)),
                                                       PauliBlock('Y',trial_2(2))],
                                       entanglement_blocks=[PauliBlock('ZZ',trial_2(2)),
                                                           PauliBlock('ZZZ',trial_2(2))],entangle_strategy='linear')
    preprocessors = [PCA(n_components=n_qubit, random_state=3) , MinMaxScaler((-1,1))]
    seed = random_seed*i
    # split the data into training set, validation set and test set
    x_train, x_val, x_test, y_train, y_val, y_test = prepare_train_val_test(x, y, train_size=n_event, 
                                                                            val_size=10, test_size=n_event, 
                                                                            preprocessors=preprocessors, random_state=seed, stratify=y)                                                                                         
    print('INFO: Training trial {} for {} qubit and {} events'.format(i, n_qubit, n_event))
    train_state_vectors = encoding_circuit.get_state_vectors(x_train)
    test_state_vectors = encoding_circuit.get_state_vectors(x_test)
    train_kernel_matrix = QSVM.get_kernel_matrix(train_state_vectors, train_state_vectors)
    train_test_kernel_matrix = QSVM.get_kernel_matrix(test_state_vectors, train_state_vectors)
    svc = svm.SVC(kernel='precomputed', probability=True, random_state=random_seed)
    tune_params = {
                   'C': [1,2,3,4,5,8,9,10,15,20,30,50,100,200,400,800,1000]
                  }
    clf = GridSearchCV(estimator=svc, param_grid=tune_params, n_jobs=-1, cv=3, scoring='roc_auc')
    clf.fit(train_kernel_matrix,y_train)
    score = clf.predict_proba(train_test_kernel_matrix)[:,1]
    roc_auc =  roc_auc_score(y_test, score)
    aucs[i] = roc_auc
print(aucs.mean())
```

```python
# load qsvm instance
backend_name = 'statevector_simulator'
n_shots = 1024
aqua_globals.random_seed = random_seed
quantum_instance = QuantumInstance(Aer.get_backend(backend_name), 
                                   shots=n_shots, 
                                   seed_simulator=random_seed, 
                                   seed_transpiler=random_seed,
                                   noise_model=None)
```

```python
# load feature map
depth = 3
entanglement = 'full'
feature_map = FeatureMapDev(n_qubit, reps=depth, entanglement=entanglement)
```

```python
feature_map.__dict__
```

```python
# load qsvm instance
batch_size = 200
qsvm = QSVM(feature_map)
qsvm.BATCH_SIZE = batch_size
```

```python
from qiskit import ClassicalRegister, QuantumCircuit
```

```python
if isinstance(feature_map, QuantumCircuit):
            use_parameterized_circuits = True
else:
    use_parameterized_circuits = feature_map.support_parameterized_circuit
```

```python
x1_vec = x_train
x2_vec=None
```

```python
if x2_vec is None:
    is_symmetric = True
    x2_vec = x1_vec
else:
    is_symmetric = False

is_statevector_sim = quantum_instance.is_statevector

measurement = not is_statevector_sim
```

```python
measurement_basis = '0' * feature_map.num_qubits
mat = np.ones((x1_vec.shape[0], x2_vec.shape[0]))
```

```python
# get all indices
if is_symmetric:
    mus, nus = np.triu_indices(x1_vec.shape[0], k=1)  # remove diagonal term
else:
    mus, nus = np.indices((x1_vec.shape[0], x2_vec.shape[0]))
    mus = np.asarray(mus.flat)
    nus = np.asarray(nus.flat)
```

```python
if is_symmetric:
    to_be_computed_data = x1_vec
else:
    to_be_computed_data = np.concatenate((x1_vec, x2_vec))
```

```python
# build parameterized circuits, it could be slower for building circuit
# but overall it should be faster since it only transpile one circuit
feature_map_params = ParameterVector('x', feature_map.feature_dimension)
parameterized_circuit = QSVM._construct_circuit(
    (feature_map_params, feature_map_params), feature_map, measurement,
    is_statevector_sim=is_statevector_sim)
parameterized_circuit = quantum_instance.transpile(parameterized_circuit)[0]
circuits = [parameterized_circuit.assign_parameters({feature_map_params: x})
            for x in to_be_computed_data]
```

```python
results = quantum_instance.execute(circuits,
                                   had_transpiled=use_parameterized_circuits)
```
