```python
import os
import glob
import numpy as np
from sklearn.metrics import auc
from sklearn.metrics import roc_curve

algorithms = ['GeneralPauli']
indir = '/eos/user/c/chlcheng/qsvm/hmumu/logs/'
outdir = '/afs/cern.ch/work/c/chlcheng/Repository/qml_work/ttH/output'

def combine_auc(data):
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)
    for d in data:
        fpr = d['fpr']
        tpr = d['tpr']
        aucs.append(auc(fpr, tpr))
        interp_tpr = np.interp(mean_fpr, fpr, tpr)
        interp_tpr[0] = 0.0
        tprs.append(interp_tpr)
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    return {'fpr':mean_fpr, 'tpr':mean_tpr, 'auc': mean_auc,'std_auc':std_auc}


data = {}
for algo in algorithms:
    data[algo] = {}
    temp = glob.glob(os.path.join(indir, algo+'*.npz'))
    for t in temp:
        d = dict(np.load(t, allow_pickle=True))
        n_feature = int(d['feature_dimension'])
        n_event = int(d['train_size']) 
        if (n_feature, n_event) not in data[algo]:
            data[algo][(n_feature, n_event)] = {'raw': []}
        fpr, tpr, _ = roc_curve(d['y_test'], d['score'])
        data[algo][(n_feature, n_event)]['raw'].append({'fpr':fpr,'tpr':tpr})
    for key in data[algo]:
        data[algo][key].update(combine_auc(data[algo][key]['raw']))

if not os.path.exists(outdir):
    os.makedirs(outdir, exist_ok=True)

    
np.savez(os.path.join(outdir, 'combined_result_qsvm.npz'), **data)
```

```python
import numpy as np 
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns

result = dict(np.load('/afs/cern.ch/work/c/chlcheng/Repository/qml_work/ttH/output/combined_result_all.npz', allow_pickle=True))
algorithms = ['qsvm_quple', 'qsvm_qiskit', 'xgboost', 'svm','dnn']
n_qubits = [20, 18, 15, 12, 10, 7, 5]
n_events = [100,200,400,800,1600, 3200]

for algo in algorithms:
    grid = np.zeros((len(n_qubits), len(n_events)))
    for i, n_qubit in enumerate(n_qubits):
        for j, n_event in enumerate(n_events):
            if (n_qubit, n_event) in result[algo].item():
                d = result[algo].item()[(n_qubit, n_event)]
                grid[i,j] = d['auc']
    print(algo)
    plt.clf()
    plt.title('{} AUC heatmap'.format(algo))
    ax = sns.heatmap(grid, xticklabels=[str(i) for i in n_events],
    	yticklabels=[str(i) for i in n_qubits], cmap="YlGnBu", mask= (grid==0))
    plt.savefig('{}_auc_heatmap.eps'.format(algo))
    plt.savefig('{}_auc_heatmap.png'.format(algo), dpi=300, bbox_inches = 'tight', pad_inches = 0)
    plt.show()
```
