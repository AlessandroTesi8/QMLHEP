```python
from quple import QuantumCircuit
from quple import ParameterisedCircuit
```

```python
a = ParameterisedCircuit(n_qubit=5, rotation_blocks=['H','RX','RY','RZ'],entanglement_blocks=['CNOT','XX'], entangle_strategy='linear',final_rotation_layer=False, copies=2)
```

```python
a
```

```python
a.flatten()
a
```

```python
a.unflatten()
a
```

```python
a.build(['H'],['CNOT'], entangle_strategy='linear', copies=2)
a
```

```python
a.add_rotation_layer(['RZ'])
a
```

```python
a.add_entanglement_layer(['XX'], entangle_strategy='full')
a
```

```python
b = ParameterisedCircuit(n_qubit=5, rotation_blocks=['RX'])
b
```

```python
a.merge(b)
a
```
